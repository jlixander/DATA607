# Load libraries and load data

```python
#import libraries
import pandas as pd
import numpy as np
import re
```

```python
#loading data
fileObject = open("7645617.txt", "r")
original_data = fileObject.read()
print(original_data)
```

```python
#Create copy of original data
data = original_data
```

# Data Wrangling

```python
#parse data into players
my_list = data.split("-----------------------------------------------------------------------------------------\n")
```

```python
#Define a function to test if number(s) are present in string
def has_numbers(str):
    return bool(re.search(r'\d', str))
```

```python
#Extract all player data into lists
plyr_num = []
plyr_name = []
plyr_state = []
plyr_tot_pts = []
plyr_pre_rnk = []
plyr_post_rnk = []
opponents = []

for player in my_list[2:]:
    info_list = player.split("|")  
    
    #Append info to list
    m1 = re.search("\d+", info_list[0])
    plyr_num.append(m1.group(0))
    
    plyr_name.append(info_list[1].strip())
    
    plyr_tot_pts.append(info_list[2].strip())
    
    plyr_state.append(info_list[10].strip())
    
    m2 = re.search("R: +(\d+).", info_list[11])
    plyr_pre_rnk.append(m2.group(1))
    
    m3 = re.search("(?=-> (\d+)|->(\d+))", info_list[11])
    if m3.group(2):
        plyr_post_rnk.append(m3.group(2))
    else:
        plyr_post_rnk.append(m3.group(1))
        
        
    #Create a list with all opponents
    opp_lst = []
    for item in info_list[3:10]:
        if has_numbers(item) == True:
            m4 = re.search("\d+", item)
            opp_lst.append(m4.group(0))
    opponents.append(opp_lst) 
```

```python
#check opponents
opponents
```

```python
pd.set_option('display.max_rows', None) #sets number of viewable rows in df
```

```python
df = pd.DataFrame(list(zip(plyr_num, 
                           plyr_name, 
                           plyr_state, 
                           plyr_tot_pts, 
                           plyr_pre_rnk, 
                           plyr_post_rnk, 
                           opponents)),
                  columns =['NUM', 
                            'NAME', 
                            'STATE', 
                            'TOT_PTS', 
                            'PRE_RNK',
                            "POST_RNK", 
                            "OPPONENTS" ]).set_index('NUM') #set 'NUM' column as index
df
```

# Get average ranking of opponents

```python
mean_opp_rnk_lst = []               #initilize empty list

for index, row in df.iterrows():    #loop through each player
    den = len(row['OPPONENTS'])      #get denominator
    running_sum = 0                  #initialize running sum
    
    for item in row['OPPONENTS']:    #get the rank of each opponent and add it the running sum
        running_sum += int(df.loc[df.index == item, 'PRE_RNK'].tolist()[0])
        
    #calculate mean oppenent rank
    mean_opp_rnk_lst.append(round(running_sum/den)) #append the sum of all rankings/number of opponents
```

```python
#Add avg opponent rank list to dataframe
df['OPP_AVG_RNK'] = mean_opp_rnk_lst
```

# Save data to .CSV file

```python
#save dataframe to csv file.
df.to_csv('output_file.csv', 
          columns=['NAME',                       #Select only the necessary columns
                    'STATE', 
                    'TOT_PTS', 
                    'PRE_RNK', 
                    'OPP_AVG_RNK'], index=False) #excludes index column
```
